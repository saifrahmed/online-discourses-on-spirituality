```python
import json
from wordcloud import (WordCloud, get_single_color_func)
import matplotlib.pyplot as plt
```

```python
INPUT_FILE__REDDIT_QUES = 'reddit_content_spirituality.json'
```

```python
titles = []
contents = []

with open(INPUT_FILE__REDDIT_QUES) as ques_file:
    data = json.load(ques_file)
    for d in data:
        if 'title' in d:
            titles.append(d['title'])
        if 'content' in d:
            contents.append(d['content'])       

blob_titles = " ".join([t for t in titles])
blob_contents = " ".join([c for c in contents])
blob_everything = blob_titles + " " + blob_contents
```

```python
def show_wordcloud(input_words):
    # Since the text is small collocations are turned off and text is lower-cased
    wc = WordCloud(collocations=False, width=1600, height=800).generate(input_words.lower())
    plt.figure( figsize=(20,10), facecolor='k')
    plt.imshow(wc, interpolation="bilinear")
    plt.axis("off")
    plt.show()
```

```python
show_wordcloud(input_words=blob_titles)
```

```python
show_wordcloud(input_words=blob_contents)
```

```python
show_wordcloud(input_words=blob_everything)
```
